```python
import openpyxl
import pandas as pd
workbook = openpyxl.Workbook()
sheet=workbook.active
data=[
    ['NAME','DOMAIN','AGE','LOCATION','SALARY','EXP'],
    ['ALEX','TESTING',25,'BNG',5000,2],
    ['MIKE','JAVA',26,'CHE',10000,3],
    ['BETA','C',30,'PUNE',15000,4],
    ['CHERRY','PYTHON',37,'MUMB',20000,5],
    ['CATHY','DS',33,'HYD',50000,6]
    ]
for row in data:
    sheet.append(row)
workbook.save('data.xlsx')
```
